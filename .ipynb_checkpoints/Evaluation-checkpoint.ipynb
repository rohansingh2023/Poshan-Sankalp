```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("datasets/Wasting.csv")
df.head()
```

```python
class MulticlassClassification(nn.Module):
    def __init__(self, num_feature, num_class):
        super(MulticlassClassification, self).__init__()
        
        self.layer_1 = nn.Linear(num_feature, 512)
        self.layer_2 = nn.Linear(512, 128)
        self.layer_3 = nn.Linear(128, 64)
        self.layer_out = nn.Linear(64, num_class) 
        
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.2)
        self.batchnorm1 = nn.BatchNorm1d(512)
        self.batchnorm2 = nn.BatchNorm1d(128)
        self.batchnorm3 = nn.BatchNorm1d(64)
        
    def forward(self, x):
        x = self.layer_1(x)
#         x = self.batchnorm1(x)
        x = self.relu(x)
        
        x = self.layer_2(x)
#         x = self.batchnorm2(x)
        x = self.relu(x)
        x = self.dropout(x)
        
        x = self.layer_3(x)
#         x = self.batchnorm3(x)
        x = self.relu(x)
        x = self.dropout(x)
        
        x = self.layer_out(x)
        
        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
model = MulticlassClassification(num_feature = len(X.columns), num_class=3)
# model.to(device)
```

```python
checkpt = torch.load('wasting_model.pth', map_location= torch.device("cpu"))
```

```python
model.load_state_dict(checkpt)
```

```python
model.eval()
```

```python
single_row = X.iloc[0,:]
type(single_row)
```

```python
sc = MinMaxScaler()

feat = sc.fit_transform(np.float32(single_row.values.reshape(-1,1)))
```

```python
input_tensor = torch.tensor(feat, dtype=torch.float32)
```

```python
with torch.no_grad():
    output = model(input_tensor)
```
