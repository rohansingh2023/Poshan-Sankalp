```python
import pandas as pd 
import numpy as np
import seaborn as sns
import pickle
from joblib import dump, load
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df = pd.read_csv('datasets/Wasting.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Wasting/Overweight'] = label_encoder.fit_transform(df['Wasting/Overweight'])
```

```python
df.head()
```

```python
X = df.drop("Wasting/Overweight", axis=1)
y = df["Wasting/Overweight"]
```

```python
len(X.columns)
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
X_train
```

```python
regressor = DecisionTreeRegressor(random_state = 0) 
```

```python
regressor.fit(X_train, y_train) 
```

```python
y_pred = regressor.predict(X_test) 
```

```python
regressor.score(X_test, y_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
def plotGraph(y_test,y_pred,regressorName):
    if max(y_test) >= max(y_pred):
        my_range = int(max(y_test))
    else:
        my_range = int(max(y_pred))
    plt.scatter(range(len(y_test)), y_test, color='blue')
    plt.scatter(range(len(y_pred)), y_pred, color='red')
    plt.title(regressorName)
    plt.show()
    return
```

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Petal Width')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.distplot(y_test-y_pred)
plt.show()
```

```python
class Prediction:
    def __init__(self, df):
        self.df = df
    def encode_target(self):
        label_encoder = LabelEncoder()
        self.df['Stunting'] = label_encoder.fit_transform(self.df['Stunting'])
    def train(self):
        X = self.df.drop("Stunting", axis=1)
        y = self.df["Stunting"]
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.30)
        regressor = DecisionTreeRegressor(random_state = 0) 
        return regressor.fit(self.X_train, self.y_train) 
    def evaluate(self):
        self.y_pred = regressor.predict(self.X_test) 
        print('MAE:', metrics.mean_absolute_error(self.y_test, self.y_pred))
        print('MSE:', metrics.mean_squared_error(self.y_test, self.y_pred))
        print('RMSE:', np.sqrt(metrics.mean_squared_error(self.y_test, self.y_pred)))
        return regressor.score(self.X_test, self.y_test)
    def plot_graph(self):
        residuals = self.y_test - self.y_pred
        plt.scatter(self.y_pred, residuals)
        plt.xlabel('Predicted Petal Width')
        plt.ylabel('Residuals')
        plt.title('Residual Plot')
        plt.axhline(y=0, color='r', linestyle='--')
        plt.show()
#     def test_eval(self):
        
```

```python
model = Prediction(pd.read_csv("datasets/Stunting.csv"))
```

```python
model.encode_target()
```

```python
model.train()
```

```python
model.evaluate()
```

```python
model.plot_graph()
```
